```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random

pred_dir = r"D:\PROJECTS\PolypExplainer\output\predictions"
gt_dir = r"D:\PROJECTS\PolypExplainer\data\raw\Kvasir-SEG\masks"

files = os.listdir(pred_dir)

# random 5 ảnh
sample_files = random.sample(files, min(5, len(files)))

for file in sample_files:
    if not file.endswith("_mask.png"):
        continue

    base_name = file.replace("_mask.png", "")
    gt_name = base_name + ".png"

    pred_path = os.path.join(pred_dir, file)
    gt_path = os.path.join(gt_dir, gt_name)

    if not os.path.exists(gt_path):
        print(f"Không tìm thấy GT cho: {file}")
        continue

    # đọc ảnh
    pred = cv2.imread(pred_path, 0)
    gt = cv2.imread(gt_path, 0)

    # resize nếu cần
    if pred.shape != gt.shape:
        gt = cv2.resize(gt, (pred.shape[1], pred.shape[0]))

    # nhị phân hóa
    pred_bin = pred > 127
    gt_bin = gt > 127

    # tạo overlay
    overlay = np.zeros((pred.shape[0], pred.shape[1], 3), dtype=np.uint8)

    overlay[(pred_bin == 1) & (gt_bin == 1)] = [0, 255, 0]   # TP
    overlay[(pred_bin == 1) & (gt_bin == 0)] = [255, 0, 0]   # FP
    overlay[(pred_bin == 0) & (gt_bin == 1)] = [0, 0, 255]   # FN

    # hiển thị
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 3, 1)
    plt.title("Prediction")
    plt.imshow(pred, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.title("Ground Truth")
    plt.imshow(gt, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.title("Overlay")
    plt.imshow(overlay)
    plt.axis('off')

    plt.suptitle(base_name)
    plt.show()
```
